```python
import numpy as np 
import pandas as pd
from sklearn.cluster import KMeans 
```

```python
#Generating random data 

np.random.seed(0)
n_students = 100
makes_obtained= np.random.randint(0,100,n_students)
overall_Score = np.random.randint(0,100,n_students)
attendance_percentage = np.random.randint(50,100,n_students)
assignment_score = np.random.randint(0,10,n_students)

overall_Score = makes_obtained*0.4 + attendance_percentage*0.4 + assignment_score*0.2
```

```python
data = pd.DataFrame({
    "makes_obtained":overall_Score,
    "attendance_percentage":attendance_percentage,
    "assignment_score":assignment_score,
    "overall_Score":overall_Score,
})
```

```python
data
```

```python
data["cluster"]=np.where(data["overall_Score"]<40,0,np.where(data["overall_Score"]<75,1,2))
```
